### HOW TO GET MULPIPLE PLOTS?

Aquí plt.subplot($m$, $n$, $k$) donde $m$ representa la cantidad de filas, $n$ la cantidad de columnas y $k$ en qué ventana graficar

```python
x=[3,4,7,9]
y=[e**2 for e in x]
y
```

```python
plt.subplot(2,2,1)
plt.plot(x,y, 'or',markersize=10)
plt.subplot(2,2,4)
plt.plot(x,y, '*k',markersize=10)

plt.show()
```

con $plot.ion()$ se pueden hacer ediciones en el gráfico sin esperar un $plt.show()$. edición "interactiva"

```python
plt.ion()
plt.subplot(3,1,1)
plt.plot(x,y, 'or',markersize=10)
plt.subplot(3,1,3)
plt.plot(x,y, '*k',markersize=10)
```

#### Transformar **Grafo Dirigido** a *Grafo NO Dirigido*

Realizamos esta transformación con el fin de encontrar las *componentes conexas* de la red génica de E.coli, la cual es un grafo dirigido.

```python
E= nx.Graph(redE)
E
```

```python
#COMPONENTES CONEXAS ---> NOT FOR DIRECTED GRAPH
nx.number_connected_components(E)
```

¿cuántos **nodos** tienen cada una de las **24 componentes conexas**?

Determina número de **grafos conexos** con determinado número de nodos

```python
E_components = nx.connected_component_subgraphs(E)
E_components
```

```python
for g in E_components:
    print(g.order())
```

En comparación con el de **GRAFO ALEATORIA DE BARABASI**, este último solo poseerá un grafo conexo, ya que cada nodo por definición estará unido a otro nodo por un solo vértice.

**AHORA**, graficamos los **GRAFOS CONEXOS**

```python
nx.draw_spring(E, node_size=100)
```

**AHORA** visualizaremos tan solo los grados presentes en el **DICCIONARIO** $E grados$ que está compuesto por todos los $nodos$ y su respectivo número de $aristas$

```python
E_grados = nx.degree(E)
type(E_grados)
```

```python
#primero extraigo los valores del diccionario,
#luego los reduzco a un conjunto con los elementos únicos -sin repeticiones- y
#finalmente los ordeno de menor a mayor
Eg = sorted(set(E_grados.values()))
len(Eg)
```

```python
Eg
```

**AHORA**, con el fin de OBTENER la FRECUENCIA de cada uno de los GRADOS obtenidos arriba, a partir de la LISTA de valores extraida del DICCIONARIO, la cual contiene todas las repeticiones de los GRADOS, **CONTAREMOS** cada una de sus repeticiones:

***-aquí será importante la LISTA DE COMPRENSIÓN-***

```python
E_freq= [E_grados.values().count(g) for g in Eg]
plt.loglog(Eg, E_freq)
plt.xlabel('grados')
plt.ylabel('frequencia x grado')
```

## UNAM: Día 2 / GENERATE A RANDOM GRAPH

Generamos un **GRAFO ALEATORIO** con nx.barabasi_albert_graph($n$,$m$) 
donde
$n$ es el número de nodos en el grafo y 
$m$ es el número de aristas de un nuevo nodo

```python
#GRAFO ALEATORIO
B= nx.barabasi_albert_graph(redE.order(),min(DEc.values())) #VALUES extracted from Ecoli GRN: #VERTICES and min(#ARISTAS)=1
B
```
